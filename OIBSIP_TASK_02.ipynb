# VISUALIZATION

```python
df_unem.columns= ["States","Date","Frequency",
               "Estimated Unemployment Rate",
               "Estimated Employed",
               "Estimated Labour Participation Rate",
               "Region","longitude","latitude"]
```

```python

plt.figure(figsize=(12, 10))
sns.heatmap(df_unem.corr())
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.lineplot('Date',y = 'Estimated Employed', data = df_unem)
plt.title("Unemployement in 2011")
plt.xticks(rotation =90)
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.boxplot(x ='Region',y = 'Estimated Employed', data = df_unem)
plt.title("State V/S Employement in 2011")
plt.xticks(rotation =90)
plt.xlabel("States")
plt.ylabel("Employment Rate")
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.barplot(x ='Region',y = 'Estimated Unemployment Rate', data = df_unem)
plt.title("State V/S Employement in 2011")
plt.xticks(rotation =90)
plt.xlabel("States")
plt.ylabel("Employment Rate")
plt.show()
```

```python
plt.title("Indian Unemployment")
sns.histplot(x="Region", hue="Date", data=df_unem)
plt.show()
```

```python

plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment")
sns.boxplot(x="Estimated Labour Participation Rate", hue="longitude", data=df_unem,color='red')
plt.show()
```

```python
plt.figure(figsize = (6,2))
sns.pairplot(df_unem)
plt.subplots_adjust(hspace =0.5)
```
